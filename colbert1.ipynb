```python
from ragatouille import RAGPretrainedModel

# Load ColBERT model (support multilingual)
RAG = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")
```

```python

# Index dokumen
passages = [doc.page_content for doc in your_documents]
metadatas = [doc.metadata for doc in your_documents]

RAG.index(
    collection=passages,
    document_ids=[str(i) for i in range(len(passages))],
    document_metadatas=metadatas,
    index_name="legal_docs",
    max_document_length=512,
    split_documents=True
)

# Search
results = RAG.search(
    query="Berapa lama batas waktu penuntutan untuk kejahatan ringan?",
    k=10
)

for i, result in enumerate(results, 1):
    print(f"[{i}] Score: {result['score']:.4f}")
    print(f"    {result['content'][:200]}\n")
```
